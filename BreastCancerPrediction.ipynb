```python
drop_list1 = ['area_mean','radius_mean','compactness_mean','concavity_mean','radius_se','perimeter_se',
              'radius_worst','perimeter_worst','compactness_worst','concavity_worst',
              'concavity_se','concave points_se','texture_mean','area_mean']
filterdf = x.drop(drop_list1,axis = 1 ) 
filterdf.head()
```

```python
f,ax = plt.subplots(figsize=(14, 14))
sns.heatmap(filterdf.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
SEED = 2
```

```python
X_train,X_test,y_train,y_test = train_test_split(filterdf, y, test_size=0.3,stratify = y,random_state = SEED)
```

```python
dt = DecisionTreeClassifier(random_state = SEED)
```

```python
dt.get_params()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params_dt = {'max_depth':[3,4,5,6],'min_samples_leaf':[0.04,0.06,0.08],'max_features':[0.2,0.4,0.6,0.8]}
```

```python
grid_dt = GridSearchCV(estimator = dt,param_grid = params_dt,scoring='accuracy',cv=10,n_jobs=-1)
```

```python
grid_dt.fit(X_train,y_train)
```

```python
best_hyperparams = grid_dt.best_params_
```

```python
print('Best Hyperparameters:\n',best_hyperparams)
```

```python
best_CV_score = grid_dt.best_score_
best_CV_score
```

```python
best_model = grid_dt.best_estimator_
best_model
```

```python
test_acc = best_model.score(X_test, y_test)
test_acc
```

```python
#DecisionTree Classification with Hyperparameter tuning gives accuracy of 91.2%
```

```python
#Performing ensemble learning algorithm on the dataset
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
lr = LogisticRegression(random_state = SEED)
```

```python
knn = KNN()
```

```python
classifiers = [('Logistic Regression',lr),('K Nearest Neighbors', knn),('Classification Tree', dt)]
```

```python
for clf_name,clf in classifiers:
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    print('{:s}:{:.3f}'.format(clf_name,accuracy_score(y_test,y_pred)))
```

```python
vc = VotingClassifier(estimators = classifiers)
```

```python
vc.fit(X_train,y_train)
```

```python
y_pred = vc.predict(X_test)
```

```python
print('Voting Classifier: {:.3f}'.format(accuracy_score(y_test,y_pred)))
```

```python
#Thus ensemble learning increases accuracy to 93%
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix

# split data train 70 % and test 30 %
x_train, x_test, y_train, y_test = train_test_split(filterdf, y, test_size=0.3, random_state=SEED)

#random forest classifier with n_estimators=10 (default)
clf_rf = RandomForestClassifier(random_state=SEED)      
clr_rf = clf_rf.fit(x_train,y_train)

ac = accuracy_score(y_test,clf_rf.predict(x_test))
print('Accuracy is: {:.3f}'.format(ac))
cm = confusion_matrix(y_test,clf_rf.predict(x_test))
sns.heatmap(cm,annot=True,fmt="d")
```
